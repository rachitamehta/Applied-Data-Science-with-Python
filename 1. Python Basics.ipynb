### Grouping

```python
datagrp=pd.DataFrame({"gender":["M","F","M","M","F"], "Age":[34,56,89,23,78], "Weight":[45,75,90,37,57]})
datagrp
```

```python
datagrp.groupby("gender")
```

```python
datagrp.groupby("gender").sum()
```

```python
datagrp.groupby("gender")["Age"].mean()
```

```python
# average profit in every outlet size

Data.groupby("Outlet_Size")["Profit"].mean()
```

```python
# average profit, item outlet sale in every outlet size

Data.groupby("Outlet_Size")["Profit","Item_Outlet_Sales"].mean()
```

```python
# sum and average of profit,item outlet sales in every outlet size

# using agg and dictionary to calculate both mean and sum for profit and item outlet sales

Data.groupby("Outlet_Size").agg({"Profit":["mean","sum"],"Item_Outlet_Sales":["mean","sum"]})
```

```python
Data.groupby(["Outlet_Size","Item_Type"])["Profit","Item_Outlet_Sales"].mean()
```

### Concatinate and Merge

```python
df1=pd.DataFrame({"Name":["A","B","C","D"],"Age":[34,78,39,70],"Weight":[47,78,36,70]})
df1
```

```python
df2=pd.DataFrame({"Name":["C","B","A","D"],"Qualification":["BE","PhD","MS","ME"],"Skill":["Python","java","C++","SQL"]})
df2
```

```python
# Concatination

# the below function adds row wise

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1)

# Data is concatinated column wise however there are 2 name columns. Concatinate doesn't check any value. 
```

```python
# Merge the data based on name 

pd.merge(df1,df2, on="Name")
```

### Append

```python
df3=pd.DataFrame({"Name":["A","B","C","D"], "Age":[34,78,39,70], "Weight":[47,78,36,70]})
df3
```

```python
df4=pd.DataFrame({"Name":["x","y","z","w"], "Age":[94,88,67,56], "Weight":[97,58,78,90]})
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4],ignore_index=True)
```

```python
# Append Function can be used instead of   pd.concat([df1,df2],ignore_index=True)

df3.append(df4, ignore_index=True)
```
