1. Line Chart

6. Count Plot
7. Box Plot

10. Swarm Plot
11. Violin Plot
12. Pair Plot

## Basic Plots using Matplotlib

### Line Chart - bivariant plot

```python
t=np.linspace(0,20,10)   # linspace = linear space 
t
```

```python
# y = f(x)
# therefore y = sin(x)

y=np.sin(t)
y
```

```python
plt.plot(t,y)
```

```python
plt.plot(t,y,color="r")
```

```python
y1=np.cos(t)
y1
```

```python
plt.plot(t,y1,color='y')
```

```python
# Plot Sin t (y) and Cos t (y1) on the same line graph

plt.plot(t,y)
plt.plot(t,y1)
plt.show()

# Plot is not showing which line is sin or cos
```

```python
# Providing Labels to the data

plt.plot(t,y, label="sinwave")
plt.plot(t,y1, label ="coswave")
plt.show()

# This data still doesn't show the labels
```

```python
# Incorporating Legend to show the labels

plt.plot(t,y, label="sinwave")
plt.plot(t,y1, label ="coswave")
plt.legend()
plt.show()
```

### Pie Chart

- Helps to understand percentage distribution
- Bivariant
- 1 numberical and 1 categorical variable

```python
agent = ["Kumar", "John", "Vani", "Jacob", "Ameen"]

sales = [24,56,47,32,18]

plt.bar(agent, sales)
```

```python
plt.pie(sales, labels = agent)
plt.show()
```

```python
plt.pie(sales, labels = agent, autopct="%.2f")
plt.show()
```

### Plots using Seaborn

```python
data = pd.read_csv("/Users/macuser/Documents/PGP AI & ML/Applied DS with Python/Datasets/hotel/hotel.csv")
data
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data
```

```python
# relation between total bill and tips

sns.lineplot(x=data["total_bill"], y=data["tip"])
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x=data["total_bill"],y=data["tip"])
```

```python
# filtering the above chart based on sex

plt.figure(figsize = (10,5))
sns.scatterplot(x=data["total_bill"],y=data["tip"], hue=data["sex"])
```

- It can be inferred that the data is not fully linear
- There are times when there is more bill there are more tips but in all cases

```python
# What is the max bill paid by gender?

sns.barplot(x=data["sex"], y=data["total_bill"])
```

```python
sns.barplot(x=data["sex"], y=data["total_bill"], ci=None)  # ci = confidence interval
```

```python
# The bar chart calculated the average value. This is a default function
# What is the max value paid by each gender?

sns.barplot(x=data["sex"], y=data["total_bill"], ci=None, estimator = np.max)  # ci = confidence interval
```

```python
# What is the max value paid by each gender?

sns.barplot(x=data["sex"], y=data["total_bill"], ci=None, estimator = np.max, hue=data["day"])
```

```python
# Count of transactions based on days

data["day"].value_counts()
```

```python
sns.countplot(data["day"])
```

```python
# histogram / distribution plot

sns.distplot(data["total_bill"])
```

```python
sns.distplot(data["total_bill"],kde=False)
```

```python
sns.distplot(data["total_bill"],hist=False)
```

```python
# Strip Plot - bivariant plot
# Strip Plot is a combination of bar chart and scatter plot
# 1 numerical and 1 categorical

sns.stripplot(x=data["sex"], y=data["total_bill"])
```

```python
sns.stripplot(x=data["sex"], y=data["total_bill"], hue = data["day"])
```

```python
# Swarm Plot - similar to Strip Plot
# Swarm Plot - data is not overlapping. Helps to differentiate the data points

sns.swarmplot(x=data["sex"], y=data["total_bill"], hue = data["day"])
```

```python
sns.swarmplot(x=data["sex"], y=data["total_bill"], hue = data["time"])
```
